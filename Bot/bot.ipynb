```python
import os
from dotenv import load_dotenv

load_dotenv()
groq_api_key = os.environ["GROQ_API_KEY"]
huggingface_api_key = os.environ["HUGGINGFACE_API_KEY"]
```

```python
from langchain_community.document_loaders import TextLoader

file_path = r"data.txt"
loader = TextLoader(file_path)
documents = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 80, chunk_overlap = 20)
docs_splitted = text_splitter.split_documents(documents)
docs_splitted
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(docs_splitted, embeddings)
retriever = vectorstore.as_retriever()

vectordb = Chroma.from_documents(docs_splitted, embeddings, persist_directory="./chroma_db")
```
